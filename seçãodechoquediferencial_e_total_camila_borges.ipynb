```python
#Grafico da seo de choque diferencial - teorico
#Autor: Camila Borges Sena
#Email para contato: camila19sena@gmail.com
import numpy as np
import matplotlib.pyplot as plt

# Definio dos parmetros
alpha = 1/137  # Constante de estrutura fina
m_mu = 0.105  # Massa do mon (GeV/c^2)
E_cm =  35   #Energia do centro de massa
s = E_cm**2 #Energia do centro de massa em GeV
factor = (alpha**2 / 4 ) # Limite superior terico
cos_theta = np.linspace(-0.8, 0.8, 100)  # Valores de cos(theta) para o terico
#cos_theta_exp = np.linspace(-0.8, 0.8, 20)  # Valores discretos para dados experimentais

print(factor)

# Modelo terico para a seo de choque diferencial
sdifferential_cross_section = (factor * (1 + cos_theta**2))*(0.389e6) #O fator de converso para nb  

print(f"dif: {sdifferential_cross_section}")

# Plotando o grfico
plt.figure(figsize=(8, 6))

# Curva terica
plt.plot(cos_theta, sdifferential_cross_section, label="Previso QED", linestyle="-", color="black")


# Configuraes do grfico
plt.xlabel(r"$\cos\theta$", fontsize=19)
plt.ylabel(r"$\frac{sd\sigma}{d\Omega}$ (GeV$^2$ $\cdot$ nb $\cdot$ ster$^{-1}$)", fontsize=19)
plt.xticks(fontsize=14)  # Aumenta o tamanho da fonte dos nmeros do eixo x
plt.yticks(fontsize=14)  # Aumenta o tamanho da fonte dos nmeros do eixo y
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend(fontsize=10)
plt.tight_layout()
plt.show()
```

```python
#Seo de choque total
#Definio dos parmetros
alpha = 1/137  # Constante de estrutura fina
m_mu = 0.105  # Massa do mon (GeV/c^2)
E_cm =  35   # Energia do centro de massa
s_values = np.linspace(150, 2500, 3000)  # Valores de s variando de 150 a 2000 GeV^2


# Calculando a seo de choque para cada valor de s
sigma_total_values = ((4 * np.pi * alpha**2) / (3 * s_values)) * (0.389e9)

print(f"sig: {sigma_total_values}")

# Plotando o grfico
plt.figure(figsize=(8, 6))
plt.plot(s_values, sigma_total_values, label="Seo de Choque Total ($_{total}$)", color="Black")

# Configuraes do grfico
plt.xlabel(r"$s$ ($\mathrm{GeV}^2$)", fontsize=19)
plt.ylabel(r"$\sigma_{total}$ (pb)", fontsize=19)
plt.xticks(fontsize=14)  # Aumenta o tamanho da fonte dos nmeros do eixo x
plt.yticks(fontsize=14)  # Aumenta o tamanho da fonte dos nmeros do eixo y
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend(fontsize=10)
plt.tight_layout()
plt.show()
```
