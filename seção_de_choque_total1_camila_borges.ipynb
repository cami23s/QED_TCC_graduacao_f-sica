```python
import numpy as np
import matplotlib.pyplot as plt

# Definio dos parmetros tericos
alpha = 1 / 137  # Constante de estrutura fina
m_mu = 0.105  # Massa do mon (GeV/c^2)
sqrt_s_exp = np.array([13.9, 22.3, 34.5, 35.0, 38.3, 43.6])  # s em GeV (experimental)
s_experimental = sqrt_s_exp**2  # s experimental em GeV^2

# Intervalo contnuo para suavizar a curva terica (ajustado para 150 a 2000 GeV)
s_smooth = np.linspace(150, 2000, 500)  # Intervalo de s em GeV

# Calcular a seo de choque terica para os valores contnuos
sigma_teorico_smooth = ((4 * np.pi * alpha**2) / (3 * s_smooth))  * (0.389e9)

# Calcular a seo de choque terica apenas nos valores experimentais
sigma_teorico = (((4 * np.pi * alpha**2) / (3 * s_experimental)) 
                 * (1 - (3/8) * (m_mu**4 / s_experimental))) * (0.389e9)

# Dados experimentais da tabela
sigma_experimental = np.array([472.7, 184.7, 73.2, 66.1, 56.4, 42.0])  # _ em pb
erro_experimental = np.array([36.0, 15.7, np.sqrt(1.5**2 + 2.6**2), 
                              np.sqrt(1.3**2 + 3.1**2), np.sqrt(4.4**2 + 3.7**2), 
                              np.sqrt(1.7**2 + 2.5**2)])  # Erro total

# Plotando os grficos
plt.figure(figsize=(10, 8))

# Grfico terico suavizado
plt.plot(s_smooth, sigma_teorico_smooth, '-', label=r"Terico ($\sigma_{total}$)", linewidth=1.5, color="black")

# Grfico experimental com erro
plt.errorbar(s_experimental, sigma_experimental, yerr=erro_experimental, fmt='o', color="red", 
             label="Experimental", capsize=5)

# Configuraes do grfico
plt.xlabel(r"$s$ ($\mathrm{GeV}^2$)", fontsize=19)
plt.ylabel(r"$\sigma_{total}$ (pb)", fontsize=19)
plt.xticks(fontsize=16)  # Aumenta o tamanho da fonte dos nmeros do eixo x
plt.yticks(fontsize=16)  # Aumenta o tamanho da fonte dos nmeros do eixo y
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend(fontsize=10)
plt.tight_layout()
plt.show()
```
