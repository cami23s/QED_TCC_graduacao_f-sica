```python
import numpy as np
import matplotlib.pyplot as plt

# Definição dos parâmetros
alpha = 1 / 137  # Constante de estrutura fina
m_mu = 0.105  # Massa do múon (GeV/c^2)
E_cm = 35  # Energia do centro de massa
s = E_cm**2  # Energia do centro de massa em GeV
factor = (alpha**2 / 4)  # Limite superior teórico
cos_theta = np.array([-0.7, -0.5, -0.3, -0.1, 0.1, 0.3, 0.5, 0.7])  # Valores médios de cos(theta) experimentais

# Dados experimentais (dσ/dΩ) e erros (em nb·ster⁻¹)
experimental_data = {
    "34.5 GeV": {
        "cross_section": np.array([7.12, 5.84, 5.46, 5.04, 5.45, 6.04, 7.38, 8.85]),
        "error": np.array([0.43, 0.36, 0.30, 0.29, 0.31, 0.32, 0.40, 0.50])
    },
    "35.0 GeV": {
        "cross_section": np.array([7.59, 6.68, 5.73, 5.60, 5.19, 5.18, 6.30, 6.06]),
        "error": np.array([0.46, 0.39, 0.30, 0.30, 0.29, 0.28, 0.35, 0.39])
    },
    "38.3 GeV": {
        "cross_section": np.array([8.23, 6.37, 4.94, 3.54, 5.04, 6.25, 8.66, 6.65]),
        "error": np.array([1.85, 1.42, 1.08, 0.88, 1.05, 1.20, 1.43, 1.62])
    },
    "43.6 GeV": {
        "cross_section": np.array([8.45, 8.32, 6.28, 3.71, 4.16, 4.91, 4.83, 5.66]),
        "error": np.array([0.96, 0.84, 0.64, 0.49, 0.51, 0.55, 0.61, 0.76])
    }
}

# Modelo teórico para a seção de choque diferencial
sdifferential_cross_section = (factor * (1 + cos_theta**2)) * (0.389e6)  # Conversão para nb

# Gráficos separados para cada conjunto experimental
for energy, data in experimental_data.items():
    plt.figure(figsize=(8, 6))
    
    # Linha teórica sem erro
    plt.plot(cos_theta, sdifferential_cross_section, 
             "-", color="black", label="Previsão QED")
    
    # Dados experimentais com erros
    plt.errorbar(cos_theta, data["cross_section"], 
                 yerr=data["error"], fmt="o", label=f"\u221a{{s}}={energy}")
    
    # Configurações do gráfico
    plt.xlabel(r"$\cos\theta$", fontsize=12)
    plt.ylabel(r"$\frac{sd\sigma}{d\Omega}$ (nb $\cdot$ ster$^{-1}$)", fontsize=12)
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.legend(fontsize=10)
    plt.tight_layout()
    plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Definição dos parâmetros
alpha = 1 / 137  # Constante de estrutura fina
m_mu = 0.105  # Massa do múon (GeV/c^2)
E_cm = 35  # Energia do centro de massa
s = E_cm**2  # Energia do centro de massa em GeV
factor = (alpha**2 / 4)  # Limite superior teórico

# Intervalo contínuo para cos(theta) para suavizar a curva teórica
cos_theta_smooth = np.linspace(-1, 1, 500)  # Mais pontos para suavizar a curva

# Valores médios experimentais de cos(theta)
cos_theta = np.array([-0.7, -0.5, -0.3, -0.1, 0.1, 0.3, 0.5, 0.7])

# Dados experimentais (dσ/dΩ) e erros (em nb·ster⁻¹)
experimental_data = {
    "34.5 GeV": {
        "cross_section": np.array([7.12, 5.84, 5.46, 5.04, 5.45, 6.04, 7.38, 8.85]),
        "error": np.array([0.43, 0.36, 0.30, 0.29, 0.31, 0.32, 0.40, 0.50])
    },
    "35.0 GeV": {
        "cross_section": np.array([7.59, 6.68, 5.73, 5.60, 5.19, 5.18, 6.30, 6.06]),
        "error": np.array([0.46, 0.39, 0.30, 0.30, 0.29, 0.28, 0.35, 0.39])
    },
    "38.3 GeV": {
        "cross_section": np.array([8.23, 6.37, 4.94, 3.54, 5.04, 6.25, 8.66, 6.65]),
        "error": np.array([1.85, 1.42, 1.08, 0.88, 1.05, 1.20, 1.43, 1.62])
    },
    "43.6 GeV": {
        "cross_section": np.array([8.45, 8.32, 6.28, 3.71, 4.16, 4.91, 4.83, 5.66]),
        "error": np.array([0.96, 0.84, 0.64, 0.49, 0.51, 0.55, 0.61, 0.76])
    }
}

# Modelo teórico para a seção de choque diferencial
sdifferential_cross_section_smooth = (
    factor * (1 + cos_theta_smooth**2)
) * (0.389e6)  # Conversão para nb

# Gráficos separados para cada conjunto experimental
for energy, data in experimental_data.items():
    plt.figure(figsize=(8, 6))
    
    # Linha teórica suavizada
    plt.plot(cos_theta_smooth, sdifferential_cross_section_smooth, 
             "-", color="black", label="Previsão QED")
    
    # Dados experimentais com erros
    plt.errorbar(cos_theta, data["cross_section"], 
                 yerr=data["error"], fmt="o", label=f"\u221a{{s}}={energy}")
    
    # Configurações do gráfico
    plt.xlabel(r"$\cos\theta$", fontsize=18)
    plt.ylabel(r"$\frac{sd\sigma}{d\Omega}$ (nb $\cdot$ ster$^{-1}$)", fontsize=18)
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.legend(fontsize=10)
    plt.tight_layout()
    plt.show()
```
