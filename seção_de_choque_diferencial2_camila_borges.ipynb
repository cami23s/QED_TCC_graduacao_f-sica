```python
#Artigo 2, PETRA

import numpy as np
import matplotlib.pyplot as plt

# Definição dos parâmetros
alpha = 1 / 137  # Constante de estrutura fina
m_mu = 0.105  # Massa do múon (GeV/c^2)
E_cm = 35  # Energia do centro de massa
s = E_cm**2  # Energia do centro de massa em GeV
factor = (alpha**2 / 4)  # Limite superior teórico

# Intervalo contínuo para cos(theta) para suavizar a curva teórica
cos_theta_smooth = np.linspace(-1, 1, 500)  # Mais pontos para suavizar a curva

# Valores atualizados de cos(theta)
cos_theta = np.array([0.70, 0.50, 0.30, 0.10, -0.10, -0.30, -0.50, -0.70])

# Dados experimentais atualizados (dσ/dΩ) e erros (em nb·ster⁻¹)
experimental_data = {
    "35.0 GeV": {
        "cross_section": np.array([7.05, 5.79, 5.43, 5.03, 5.46, 6.07, 7.43, 8.85]),
        "error": np.array([0.43, 0.36, 0.30, 0.29, 0.31, 0.32, 0.40, 0.50])
    }
}

# Modelo teórico para a seção de choque diferencial
sdifferential_cross_section_smooth = (
    factor * (1 + cos_theta_smooth**2)
) * (0.389e6)  # Conversão para nb

# Gráfico atualizado
plt.figure(figsize=(8, 6))

# Linha teórica suavizada
plt.plot(cos_theta_smooth, sdifferential_cross_section_smooth, 
         "-", color="black", label="Previsão QED")

# Dados experimentais com erros
plt.errorbar(cos_theta, experimental_data["35.0 GeV"]["cross_section"], 
             yerr=experimental_data["35.0 GeV"]["error"], fmt="o", label="√{s}=35.0 GeV")

# Configurações do gráfico
plt.xlabel(r"$\cos\theta$", fontsize=18)
plt.ylabel(r"$\frac{sd\sigma}{d\Omega}$ (nb $\cdot$ ster$^{-1}$)", fontsize=18)
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend(fontsize=10)
plt.tight_layout()
plt.show()

```
